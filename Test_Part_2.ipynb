# Importing Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(9,5)
```

# Loading Dataset

```python
raw_data= pd.read_csv(r"C:\Users\DELL\Desktop\juniorML_test-20221016T064540Z-001\juniorML_test\part_2\epg_data-2021-12-14.csv", sep=';', header=None)
headers = raw_data.iloc[0]
raw_data= pd.DataFrame(raw_data.values[1:], columns=headers)
raw_data.head(20)
```

# Checking NAN values

```python
raw_data.isnull().sum()
```

## Dropping subtitles because 90k rows have NaN values dropna would drop most of the data hence dropping the colomun to save dataset 

# Dropping unnecessary Columns

```python

#Dropping Uncessry Columns
raw_data=raw_data[['id', 'titles', 'episode_numbers', 'start_dt', 'end_dt',
       'computed_duration_min', 'categories',
        'channel', 'diffusion_country']]

#Dropping descriptions.lang descriptions.text and credits name as not mentioned in document attributes.
```

```python
raw_data.head(5)
```

# Dropping NaN values

```python
raw_data.dropna(inplace=True)
raw_data.isnull().sum()
```

# Converting Episode Duration from object to Float

```python
raw_data_c=raw_data['computed_duration_min']
raw_data['computed_duration_min'] = pd.to_numeric(raw_data_c)
```

# Fixing Titles and Categories (List to Str) 

```python
raw_data['titles'] = raw_data.titles.apply(lambda x: x[1:-1].split(','))
raw_data.titles= raw_data['titles'].apply(lambda x: ','.join(map(str, x)))
raw_data.titles=raw_data.titles.apply(lambda x: x.replace('"', ''))
raw_data['categories'] = raw_data.categories.apply(lambda x: x[1:-1].split(','))
raw_data.categories= raw_data['categories'].apply(lambda x: ','.join(map(str, x)))
raw_data.categories=raw_data.categories.apply(lambda x: x.replace('"', ''))
```

```python
raw_data
```

# Extracting Time of Airing

```python

raw_data['start_dt']= raw_data.start_dt.apply(lambda x:x[1:-1][10:18])
raw_data['end_dt']=raw_data['start_dt']= raw_data.end_dt.apply(lambda x:x[1:-1][10:18])
```

# Clean Dataset

```python
clean_data=raw_data
clean_data.head(20)
```

# Top 5 Shows
